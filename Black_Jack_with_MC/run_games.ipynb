# Blackjack Game
Write several blackjack algorithms using Monte Carlo methods. \
Plot value function for optimal Q-value.

```python
import torch
import gym

from mcblackjack import MCBJDetermAgent\
                      , MCBJEpsGreedyOnPolicyAgent\
                      , MCBJEpsGreedyOffPolicyAgent \
                      , mc_control_on_policy_first_visit\
                      , mc_control_on_policy_every_visit\
                      , mc_control_off_policy_first_visit \
                      , mc_control_off_policy_every_visit \
                      , mc_control_per_decision_first_visit \
                      , mc_control_per_decision_every_visit \
                      , plot_blackjack_value
```

```python
gamma = 1
n_episode = 500000

env = gym.make("Blackjack-v1")
agent = MCBJDetermAgent(env, gamma)
optimal_agent = mc_control_on_policy_first_visit(env, agent, gamma, n_episode)
env.close()


V = optimal_agent.get_Q().max(-1).values
plot_blackjack_value(V[12:22, 1:11, :])
```
